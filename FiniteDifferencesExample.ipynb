```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-poster")
```

```python
n = 10
h = (5-0) / n

# Get A
A = np.zeros((n+1, n+1))
A[0, 0] = 1
A[n, n] = 1
for i in range(1, n):
    A[i, i-1] = 1
    A[i, i] = -2
    A[i, i+1] = 1

print(A)
```

```python
# Get b
b = np.zeros(n+1)
b[1:-1] = -9.8*h**2
b[-1] = 50
print(b)
```

```python
# solve the linear equations
y = np.linalg.solve(A, b)

t = np.linspace(0, 5, 11)
```

```python
plt.figure(figsize=(10,8))
plt.plot(t, y)
plt.plot(5, 50, 'ro')
plt.xlabel('time (s)')
plt.ylabel('altitude (m)')
plt.show()
```
